#한 화면에 여러 개 그래프 나타내기

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
fig, axs= plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

#실습문제

```python
data1 = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [10, 20, 15, 25, 30],
    'category': ['A', 'B', 'A', 'B', 'A']
})
sns.scatterplot(x='x', y='y', hue='category', data=data1)
plt.show()
```

"category" 열을 사용하여 각 카테고리별로 데이터의 개수를 세어서 막대 그래프로 시각화하세요.

```python
data2 = pd.DataFrame({
'category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B'],
'value': [10, 20, 15, 25, 30, 12, 18, 22, 28, 35]
})

sns.countplot(x='category',hue='category',data=data2)
```
