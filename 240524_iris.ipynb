```python
#데이터 파일 읽어오기
import seaborn as sns
df = sns.load_dataset('iris')
```

```python
#데이터의 기본 정보 출력
df.info()
```

```python
#데이터의 기초 통계량 출력
df.describe()
```

```python
#품종별 개수 구하기
df.value_counts(['species'])
```

```python
#결측치 확인하기
df.isnull().sum()
```

```python
#중복데이터 확인하기
df[df.duplicated()]
```

```python
#중복데이터 모두 확인
idx = (df['sepal_length'] == 5.8) & (df['petal_width'] == 1.9)
df.loc[idx, :]
```

```python
#중복데이터 삭제하기
df2=df.drop_duplicates()
```

```python
df2.info()
```

```python
#붓꽃 데이터 그룹핑
grouped_sum=df2.groupby('species').sum()
grouped_sum

grouped_avg=df2.groupby('species').mean()
grouped_avg
```

```python
import matplotlib.pyplot as plt
plt.bar(df.index, df.sepal_length)
```

```python
plt.hist('sepal_length')
plt.hist('sepal_width')
plt.hist('petal_length')
plt.hist('petal_width')
```

```python
df3=df.groupby('species').mean()
plt.bar(df3.index,df3['sepal_length'])
```
